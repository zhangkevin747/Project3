```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
female_activities = pd.read_csv("Mouse_Data_Student_Copy.xlsx - Fem Act.csv")
female_temp = pd.read_csv("Mouse_Data_Student_Copy.xlsx - Fem Temp.csv")
male_activities = pd.read_csv("Mouse_Data_Student_Copy.xlsx - Male Act.csv")
male_temp = pd.read_csv("Mouse_Data_Student_Copy.xlsx - Male Temp.csv")
```

```python
female_activities['Minute Since 0'] = female_activities.index + 1
female_activities['Day'] = female_activities['Minute Since 0'] // 24 // 60 + 1
female_activities['Minute'] = female_activities['Minute Since 0'] % (24 * 60)
female_activities
```

```python
female_temp['Minute Since 0'] = female_temp.index + 1
female_temp['Day'] = female_temp['Minute Since 0'] // 24 // 60 + 1
female_temp['Minute'] = female_temp['Minute Since 0'] % (24 * 60)
female_temp
```

```python
male_activities['Minute Since 0'] = male_activities.index + 1
male_activities['Day'] = male_activities['Minute Since 0'] // 24 // 60 + 1
male_activities['Minute'] = male_activities['Minute Since 0'] % (24 * 60)
male_activities
```

```python
male_temp['Minute Since 0'] = male_temp.index + 1
male_temp['Day'] = male_temp['Minute Since 0'] // 24 // 60 + 1
male_temp['Minute'] = male_temp['Minute Since 0'] % (24 * 60)
male_temp
```

```python
male_activities = male_activities.rename(
    columns={col: f"{col}_act" for col in male_activities.columns if col != "Minute Since 0" and col != "Day" and col != "Minute"}
)
male_temp = male_temp.rename(
    columns={col: f"{col}_temp" for col in male_temp.columns if col != "Minute Since 0" and col != "Day" and col != "Minute"}
)

male_data = pd.merge(male_activities, male_temp, on=["Minute Since 0", 'Day', 'Minute'])
male_data
```

```python
female_activities = female_activities.rename(
    columns={col: f"{col}_act" for col in female_activities.columns if col != "Minute Since 0" and col != "Day" and col != "Minute"}
)
female_temp = female_temp.rename(
    columns={col: f"{col}_temp" for col in female_temp.columns if col != "Minute Since 0" and col != "Day" and col != "Minute"}
)

female_data = pd.merge(female_activities, female_temp, on=["Minute Since 0", 'Day', 'Minute'])
female_data
```

```python
data = pd.merge(female_data, male_data, on=["Minute Since 0", 'Day', 'Minute'])
data
```

```python
# Ensure EstrusStage column exists (replace 'day' if needed)
if 'EstrusStage' not in data.columns:
    data['EstrusStage'] = data['Day'] % 4  # Assuming a repeating 4-day cycle

# Compute mean activity for each female across the cycle
female_activity_cols = [col for col in data.columns if "f" in col and "_act" in col]
data["Mean_Female_Activity"] = data[female_activity_cols].mean(axis=1)

data["Hour"] = data["Minute"] // 60  # Convert minutes to hours

# Group by EstrusStage and compute mean activity
estrus_activity = data.groupby(["EstrusStage", 'Minute', 'Hour'])["Mean_Female_Activity"].mean().reset_index()
estrus_activity
```

```python
# Pivot using pivot_table (aggregating Mean_Female_Activity)
heatmap_data = estrus_activity.pivot_table(
    index="Hour", columns="EstrusStage", values="Mean_Female_Activity", aggfunc="mean"
)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap="coolwarm", annot=True, fmt=".1f")

plt.title("Average Female Activity Across Estrus Cycle (Hourly)")
plt.xlabel("Estrus Stage (0-3)")
plt.ylabel("Hour of the Day")
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
```

```python
heatmap_data = heatmap_data.reset_index()
```

```python
json_data = heatmap_data.to_json(orient="records")

# Print the JSON data
print(json_data)

# If you want to save the JSON data to a file:
with open("heatmap_data.json", "w") as file:
    file.write(json_data)
```
